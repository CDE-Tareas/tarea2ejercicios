```python
# read web table into pandas DF
import pandas as pd

linkToFile='https://github.com/CienciaDeDatosEspacial/code_and_data/raw/main/data/carbonEmi_downloaded.csv'
carbon=pd.read_csv(linkToFile)
```

```python
carbon
```

```python
#1. Keep the columns name, value, date_of_information and region.
carbon=carbon.drop(columns=["slug", "ranking"])
carbon
```

```python
#2. Change the column name date_of_information to carbon_date.
carbon=carbon.rename(columns={"date_of_information": "carbon_date"})
carbon
```

```python
#3.1 Make sure the cells with text does not have neither trailing nor leading spaces. Para 'name' 
carbon['name']=carbon['name'].str.strip()
carbon
```

```python
#3.2 Make sure the cells with text does not have neither trailing nor leading spaces. Para 'region' 
carbon['region']=carbon['region'].str.strip()
carbon
```
